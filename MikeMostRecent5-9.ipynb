<br><br>
***
**Features**:
<br><br>
##### Fiscal Year Released:
- Fiscal year (year ending June 30) for which the offender was released from prison.
<br><br>

##### Recidivism Reporting Year:
- Fiscal year (year ending June 30) that marks the end of the 3-year tracking period. For example, offenders exited prison in FY 2012 are found in recidivism reporting year FY 2015.
<br><br>

##### Race-Ethnicity: 
- Offender's Race and Ethnicity
<br><br>

##### Age At Release:
- Offender's age group at release from prison.
<br><br>

##### Convicting Offense Classification: 
- Maximum penalties:
    - *A Felony* : Life 
    - *B Felony* : 25 or 50 years 
    - *C Felony* : 10 years 
    - *D Felony* : 5 years
    - *Aggravated Misdemeanor* : 2 years
    - *Serious Misdemeanor* : 1 year
    - *Simple Misdemeanor* : 30 days
<br><br>

##### Convicting Offense Type: 
- General category for the most serious offense for which the offender was placed in prison.
<br><br>

##### Convicting Offense Subtype: 
-Further classification of the most serious offense for which the offender was placed in prison.
<br><br>

##### Release Type: 
- Reasoning for Offender's release from prison.
<br><br>

##### Main Supervising District:
- The Judicial District supervising the offender for the longest time during the tracking period.
<br><br>

##### Recidivism - Return to Prison:	
- *No* : No Recidivism
- *Yes* : Prison admission for any reason within the 3-year tracking period
<br><br>

##### Days to Recidivism:
- Number of days it took before the offender returned to prison.
<br><br>

##### New Conviction Offense Classification: 
- New conviction maximum penalties: 
    - *A Felony* : Life
    - *B Felony* : 25 or 50 years
    - *C Felony* : 10 years
    - *D Felony* : 5 years
    - *Aggravated Misdemeanor* : 2 years
    - *Serious Misdemeanor* : 1 year
    - *Simple Misdemeanor* : 30 days
<br><br>

##### New Conviction Offense Type: 
- General category for the new conviction while the offender is out of prison.
<br><br>

##### New Conviction Offense Sub Type: 
- Further classification of the new conviction.
<br><br>

##### Part of Target Population: 
- The Department of Corrections has undertaken specific strategies to reduce recidivism rates for prisoners who are on parole.
<br><br>

##### Recidivism Type:
- There is no description of this feature. It pertains only to recidivists and to the category of recidivsm they follow into.
<br><br>

##### Sex:
- Gender of our offender

#### *crime_class*

Create new column *enhance*, drop enhanced felonies from *crime_class* and put them in appropriate label group.  Add values to *enhance* for each "enhancement" required to get felony for that crime based on type and subtype and Iowa gov penal codes and sentencing guidlines.

```python
prison['enhanced'] = 0
# Enhancemets
classname = (prison.crime_class == 'C Felony')
subname = (prison.crime_subtype == 'Sex Offender Registry/Residency')
prison.loc[subname,'enhanced'] += 1
prison.loc[prison.crime_subtype == 'Robbery', 'enhanced'] += 1
prison.loc[prison.crime_subtype == 'Weapons', 'enhanced'] += 1

# Felony - Enhannced, all rows get a value in new column
classname = (prison.crime_class == 'Felony - Enhanced')
prison.loc[classname,'enhanced'] += 1

# Drug Trafficking C
subname = (prison.crime_subtype == 'Trafficking')
prison.loc[classname & subname, 'crime_class'] = 'C Felony'

# Arson goes to C 
subname = (prison.crime_subtype == 'Trafficking')
prison.loc[classname & subname, 'crime_class'] = 'C Felony'

# Murder B
subname = (prison.crime_subtype == 'Murder/Manslaughter')
prison.loc[classname & subname, 'crime_class'] = 'B Felony'

# The Rest go to D
prison.loc[classname, 'crime_class'] = 'D Felony'
```

```python
# Other "Felony (Old Code)" goes to B
classname = (prison.crime_class == 'Other Felony (Old Code)')
subname = (prison.crime_subtype == 'Murder/Manslaughter')
prison.loc[classname & subname, 'crime_class'] = 'B Felony'
```

```python
# "Felony Enhancement - To Original Penalty"
classname = (prison.crime_class == 'Felony - Enhancement to Original Penalty')

# Everything in this category has been enhanved by habitual unlawfulness  or other problems
prison.loc[classname,'enhanced'] += 1

# Drug Trafficking C
subname = (prison.crime_subtype == 'Trafficking')
prison.loc[classname & subname, 'crime_class'] = 'C Felony'

# Robbery  goes to C
subname = (prison.crime_subtype == 'Robbery')
prison.loc[classname & subname, 'crime_class'] = 'C Felony'

# Violent Murder/Manslaughter goes to B
subname = (prison.crime_subtype == 'Murder/Manslaughter')
prison.loc[classname & subname, 'crime_class'] = 'B Felony'

# The rest go to D and will all be mapped at the end
prison.loc[classname, 'crime_class'] = 'D Felony'
```

```python
# 'Other Felony' will get put in corresponding category 
classname = (prison.crime_class == 'Other Felony')

# Drug Trafficers get put in the mode category, C
subname = (prison.crime_subtype == 'Trafficking')
prison.loc[classname & subname, 'crime_class'] = 'C Felony'

# Murderers go to 'B Felony'
subname = (prison.crime_subtype == 'Murder/Manslaughter')
prison.loc[classname & subname, 'crime_class'] = 'B Felony'

# Violent Sex offender go in 'C Felony'
subname = (prison.crime_subtype == 'Sex')
prison.loc[classname & subname, 'crime_class'] = 'C Felony'

# "Sexual Predator Community Supervision" goes to Special Sentence 2005 
classname = (prison.crime_class == 'Sexual Predator Community Supervision')
prison.loc[classname, 'crime_class'] = 'Special Sentence 2005'
```
