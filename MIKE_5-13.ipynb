#### *crime_class*  & *enhanced*

Create new column *enhance*, drop enhanced felonies from *crime_class* and put them in appropriate label group.  Add values to *enhanced* for each "enhancement" required to get felony for that crime based on type and subtype and Iowa gov penal codes and sentencing guidlines.

The logic behind this involved getting a thorough understanding of the Iowa Penal Code and sentencing guidlines. Steadfast rules for sentencing are not as many as one may think, however based laws regarding the seriousness of a crime, habitual offennces and other unique characteristics, a lot can be inferred. For example, drug possession for a first offense is a low level misdemeanor and the second offense is an aggravated misdemeanor. So, it takes three arressts for that crime to be punishable as a D Felony.  That is a form of an enhancement that carries more severe punishments.  


Crimes involving weapons, robberies, and others are all enhanceable beyond the regular sentencing standards.  Similarly, every row in the *crime_class* column labled with "Enhanced" or "Felony Enhancement" falls in the category of enhanced. By studying the penal codes and carefully inspecting each crime, some firm conclusions can be reached.  A big advantage of this, is that habitual offenders fall into this category. Logically, if somebody had committed multiple crimes in the past, there is a good chance that they will do it again if they get released from prison. Creating this new column, would hopefully give another useable variable to the models.

Initial gridsearch is done inside one of three functions:
- *random_pipe* : Fits training set to pipeline with PCA and scaler of 'choice(default=StandardScaler) and performs a randomized grid search with cross validation on how every many models with corresponding parameter dictionaries passed to it. Wraps Sci-kit Learn PipeLine and RandomizedSearchCV classes. Returns a dictionary containg all the best estimators, estimator parameters, and test scores. (helper to *compare_pipes*)

<br><br>
- *search_pipe* : Identical to *random_pipe* with the acception of the gridsearch object.  This Wraps Sci-kit Learn's GridSearchCV class as well as Pipeline. Outputs a dictionary of all best estimators, best parameters stats and more. (helper to *compare_pipes*)

<br><br>
- *compare_pipes*.  Similar To above functions, but uses both, can handle fitting, gridsearching and cross validating multiple models.  Displays a Dataframe of best models with its best train and test accuracy scores.  Returns a dictionary holding all best models, parameters and scores.  Overkill for many applilcations, but very useful for comparing model performance and hyper tuning at the same time.

```python
class ModelOptimizer:
    best_score = None
    opt = None
    
    def __init__(self, model, X_train, y_train, 
                 categorical_columns_indices=None, n_fold=3, seed=2405, plot=True,
                 early_stopping_rounds=30, is_stratified=True, is_shuffle=True):
        self.model = model
        self.X_train = X_train
        self.y_train = y_train
        self.categorical_columns_indices = categorical_columns_indices
        self.n_fold = n_fold
        self.seed = seed
        self.early_stopping_rounds = early_stopping_rounds
        self.is_stratified = is_stratified
        self.is_shuffle = is_shuffle
        self.plot = plot
        
        
    def update_model(self, **kwargs):
        for k, v in kwargs.items():
            setattr(self.model, k, v)
            
    def evaluate_model(self):
        eval_metric='AUC'
        pass
    
    def optimize(self, param_space, max_evals=10, n_random_starts=10):
        start_time = time.time()
        
        @use_named_args(param_space)
        def _minimize(**params):
            self.model.set_params(**params)
            return self.evaluate_model()
        
        opt = gp_minimize(
                    _minimize,
                    param_space,
                    n_calls=max_evals, 
                    n_random_starts=n_random_starts, 
                    random_state=777, 
                    n_jobs=-1
        )
        
        best_values = opt.x
        optimal_values = dict(zip([param.name for param in param_space], best_values))
        best_score = opt.fun
        self.best_score = best_score
        self.opt = opt
        
        print(f'optimal_parameters: {optimal_values}\noptimal score: {best_score}')
        print(f'\noptimization time: {time.time() - start_time}')
        print('updating model with optimal values')
        
        self.update_model(**optimal_values)
        plot_convergence(opt)
        return optimal_values


class CatboostOptimizer(ModelOptimizer):
    def evaluate_model(self):
        validation_scores = catboost.cv(
                                catboost.Pool(
                                    self.X_train, 
                                    self.y_train, 
                                    cat_features=self.categorical_columns_indices
                                ),
                                self.model.get_params(), 
                                nfold=self.n_fold,
                                stratified=self.is_stratified,
                                seed=self.seed,
                                early_stopping_rounds=self.early_stopping_rounds,
                                shuffle=self.is_shuffle,
                                plot=self.plot
        )
        
        self.scores = validation_scores
        test_scores = validation_scores.iloc[:, 2]
        best_metric = test_scores.max()
        return 1 - best_metric
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues, 
                          print_matrix=True):
    #Check if Normalization Option is Set to True. If so, normalize the raw confusion matrix before visualizing
    #Other code should be equivalent to your previous function
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
```
